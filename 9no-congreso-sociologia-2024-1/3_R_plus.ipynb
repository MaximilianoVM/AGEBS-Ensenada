```r
# Cargar librerías
library(tidyverse)
# Cargar librerías
library(tidyverse)
library(sf)
library(ggplot2)

# Cargar datos
data10 <- read.csv("DATA/FULL_ENS_AGEBS_10.csv")
data20 <- read.csv("DATA/FULL_ENS_AGEBS_20.csv")
```

```r
# Explorar datos
glimpse(data10)
glimpse(data20)
summary(data10)
summary(data20)
```

```r
# Unir datos de 2010 y 2020
data_combined <- inner_join(data10, data20, by = "AGEB", suffix = c("_2010", "_2020"))
```

```r
# Crear nuevas variables con mutate
data_combined <- data_combined %>%
  mutate(
    cambio_poblacion = (P_TOTAL_2020 - P_TOTAL_2010) / P_TOTAL_2010,
    cambio_ninos = (P_0A5_2020 - P_0A5_2010) / P_0A5_2010,
    cambio_adultos_mayores = (POB65_MAS_2020 - POB65_MAS_2010) / POB65_MAS_2010,
    cambio_educacion = (GRAPROES_2020 - GRAPROES_2010) / GRAPROES_2010,
    cambio_hijos = (PROM_HNV_2020 - PROM_HNV_2010) / PROM_HNV_2010,
    cambio_viviendas = (VIVTOT_2020 - VIVTOT_2010) / VIVTOT_2010,
    cambio_ocupantes = (PROM_OCUP_2020 - PROM_OCUP_2010) / PROM_OCUP_2010,
    cambio_computadora = (VPH_PC_2020 - VPH_PC_2010) / VPH_PC_2010
  )
```

```r
# Función para graficar distribuciones
plot_distribution <- function(data, variable, title) {
    ggplot(data, aes(x = !!sym(variable))) +
        geom_histogram(binwidth = 0.1, fill = "blue", alpha = 0.7, boundary = 0) +
        geom_vline(aes(xintercept = mean(!!sym(variable), na.rm=TRUE)),
                             color = "red", linetype = "dashed", size = 1) +
        geom_vline(aes(xintercept = median(!!sym(variable), na.rm=TRUE)),
                             color = "green", linetype = "dashed", size = 1) +
        labs(title = title, x = variable, y = "Frequency") +
        theme_minimal()
}

# Variables a visualizar
variables <- c("P_TOTAL_2010", "P_TOTAL_2020", "POB65_MAS_2010", "POB65_MAS_2020",
                             "GRAPROES_2010", "GRAPROES_2020", "PROM_HNV_2010", "PROM_HNV_2020",
                             "VIVTOT_2010", "VIVTOT_2020", "PROM_OCUP_2010", "PROM_OCUP_2020",
                             "VPH_PC_2010", "VPH_PC_2020")

#variables <- c("GRAPROES_2020")

# Graficar distribuciones
plots <- lapply(variables, function(var) {
    plot_distribution(data_combined, var, paste("Distribución de", var))
})

# Mostrar gráficos
gridExtra::grid.arrange(grobs = plots, ncol = 2)
```

```r
# Función para graficar valores atípicos usando boxplot
plot_boxplot <- function(data, variable, title) {
  ggplot(data, aes_string(x = "1", y = variable)) +
    geom_boxplot(fill = "blue", alpha = 0.7) +
    labs(title = title, x = "", y = variable) +
    theme_minimal()
}

# Graficar boxplots
boxplots <- lapply(variables, function(var) {
  plot_boxplot(data_combined, var, paste("Boxplot de", var))
})

# Mostrar boxplots
gridExtra::grid.arrange(grobs = boxplots, ncol = 2)
```
